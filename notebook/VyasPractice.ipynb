```python
def read_all_csvs(folder):
    files = {}
    for f in os.listdir(data_folder):
        if 'csv' in f:
            files[f] = pd.read_csv(data_folder + f)

    for f in files.keys():
        obj_name = f.split('.csv')[0]
        cmd = "global {}; {} = files['{}']".format(obj_name, obj_name, f)
        exec(cmd, globals(), locals())
```
